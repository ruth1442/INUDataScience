```python
# 2. (1),(2)
from collections import Counter
document = ["A","I","expert","excellent","excellent"]
c = Counter(document)
print("---2번 (1)---")
for w, c in c.most_common(5):
    print(w,c)

print("---2번 (2)---")
document = ["A","I","expert","excellent","excellent"]
c = Counter(document)
for w, c in c.most_common(1):
    print(w,c)
```
