```python
#Lab 4.1
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
irisdf = pd.DataFrame(iris.data,columns=iris.feature_names)
irisdf['target'] = iris.target
irisdf['target'] = irisdf['target'].map({0:'setosa', 1:"versicolor", 2:"virginica"})

SL = irisdf.iloc[:,0]
SW = irisdf.iloc[:,1]
PL = irisdf.iloc[:,2]
PW = irisdf.iloc[:,3]
name = irisdf.iloc[:,4]

#꽃받침의 길이와 꽃잎의 넓이 관계성 시각화
plt.scatter(SL,PW,c=iris.target)
```

```python
#Lab 4.1
#꽃잎의 길이와 꽃받침의 넓이 관계성 시각화
plt.scatter(PL,SW,c=iris.target)
```

```python
#Lab 4.2
import numpy as np

#꽃 종류별로 데이터셋 분리
setosa = irisdf.iloc[:50,:]
versicolor = irisdf.iloc[50:100,:]
virginica = irisdf.iloc[100:150,:]

#꽃 종류별 특징들 평균
setM = np.mean(setosa, axis=0)
verM = np.mean(versicolor, axis=0)
virM = np.mean(virginica, axis=0)
irisMerge = pd.DataFrame({'setosa':setM,'versicolor':verM,'virginica':virM})
#꽃잎과 꽃받침정보가 x축인 그래프
ax = irisMerge.plot(kind='bar', title='iris data mean', figsize=(12, 4), legend=True, fontsize=12)
ax.legend(['set', 'ver','vir'], fontsize=12)
```

```python
#Lab 4.2
irisMerge = irisMerge.transpose()
#꽃 종류가 x축인 그래프
ax = irisMerge.plot(kind='bar', title='iris data mean', figsize=(12, 4), legend=True, fontsize=12)
ax.legend(['SL', 'SW','PL', 'PW'], fontsize=12)

#특징을 잘 나타내는 그래프는 아래에 표시된 꽃 종류가 x축에 들어간 그래프이다.
#꽃 종류 각각의 특징을 보기 편하기 때문이다.
#반면 꽃 종류마다 특징을 비교하기에는 위에 있는 x축에 특징이 들어간 그래프가 편하다.
```

```python
#Lab 4.3

#꽃 종류별로 데이터셋 분리
setosa = irisdf.iloc[:50,:]
versicolor = irisdf.iloc[50:100,:]
virginica = irisdf.iloc[100:150,:]

#1/4, 3/4로 데이터셋 분리
set1 = irisdf.iloc[:12,:]
set3 = irisdf.iloc[12:50,:]
ver1 = irisdf.iloc[50:63,:]
ver3 = irisdf.iloc[63:100,:]
vir1 = irisdf.iloc[100:113,:]
vir3 = irisdf.iloc[113:150,:]
iris14 = pd.concat([set1,ver1,vir1],axis=0)
iris34 = pd.concat([set3,ver3,vir3],axis=0)

#csv 파일로 저장하기
iris14.to_csv('test_dataset.csv')
iris34.to_csv('train_dataset.csv')
```

```python
#Lab 4.4

#csv 다시 불러오기
iris14 = pd.read_csv('test_dataset.csv')
iris34 = pd.read_csv('train_dataset.csv')

#각각 평균내기
#1/4 데이터셋 꽃별로 나눠서 평균
set1 = iris14.iloc[:12,:4]
ver1 = iris14.iloc[12:25,:4]
vir1 = iris14.iloc[25:38,:4]
#꽃 종류별 특징들 평균
set1M = np.mean(set1, axis=0)
ver1M = np.mean(ver1, axis=0)
vir1M = np.mean(vir1, axis=0)
iris1M = pd.DataFrame({'setosa':set1M,'versicolor':ver1M,'virginica':vir1M})
iris1M = iris1M.transpose()
#꽃 종류가 x축인 그래프
ax = iris1M.plot(kind='bar', title='iris 1/4 data mean', figsize=(12, 4), legend=True, fontsize=12)
ax.legend(['SL', 'SW','PL', 'PW'], fontsize=12)
```
