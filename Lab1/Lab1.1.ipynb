```python
import csv

a = ["one","two","three"]
b = ["four","five","six"]
c = ["seven","eight","nine"]
a.extend(b)
a.extend(c)
for i in range(0,len(a),1):
    print("%s : %s" %(i,a[i]))

f = open('lab1.1.csv', 'w',newline='') 
write = csv.writer(f)   
write.writerow(a) 
f.close()

f = open('lab1.1.csv','r')
rdr = csv.reader(f)
for line in rdr:
    print(line)
f.close()
```
