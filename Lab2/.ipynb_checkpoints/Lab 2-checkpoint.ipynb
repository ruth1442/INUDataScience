```python
#Lab 2.1 미분식 계산
import numpy as np
import matplotlib.pylab as plt
import math

#미분식이 f(x+h) - f(x-h) / (2*h) 일 때
def diff(f, x):
      h = 1e-10  # 0.0000000001
      return (f(x+h) - f(x-h)) / (2*h)

#계산해야하는 예제 식
def f1(x):
    return x**2
def f2(x):
    return x**3 + x**2 + x 
def f3(x):
    return x*np.sin(x*x*x+3)
def f4(x):
    return np.cos(np.sqrt(math.e**x + 1) / 2)

#미분값 출력하기
print(diff(f1, 2))
print(diff(f2, 2))
print(diff(f3, 2))
print(diff(f4, 2))
```
